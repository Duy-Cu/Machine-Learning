## Reading each sheet separately

```python
import pandas as pd
from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
plt.style.use('dark_background')

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

df_pg1 = pd.read_excel('Folds5x2_pp_xls_format.xls', 'Sheet1')
df_pg2 = pd.read_excel('Folds5x2_pp_xls_format.xls', 'Sheet2')
df_pg3 = pd.read_excel('Folds5x2_pp_xls_format.xls', 'Sheet3')
df_pg4 = pd.read_excel('Folds5x2_pp_xls_format.xls', 'Sheet4')
df_pg5 = pd.read_excel('Folds5x2_pp_xls_format.xls', 'Sheet5')
```

## Statistical analysis of each sheet

```python
# since all 5 sheets have same data entries but shuffled, 
# we can expect the same statistical summary
df_pg1.describe()
```

## Plotting the data frame

### Histogram plot

```python
# plotting historgram for 'Sheet1'
df_pg1.hist(bins=25, figsize=(6,5))
```

### Scatter plot

```python
# since we are interested in the hourly electrical energy output, 
# we will set this as our y-axis

df_pg1.plot(kind='scatter', x ='AT', y='PE', figsize=(16,8))
df_pg1.plot(kind='scatter', x ='V', y='PE', figsize=(16,8))
df_pg1.plot(kind='scatter', x ='AP', y='PE', figsize=(16,8))
df_pg1.plot(kind='scatter', x ='RH', y='PE', figsize=(16,8))
```

### Box plot

```python
df_pg1.boxplot(column=['AT', 'V', 'RH'] ,figsize=(16,8))
```

```python
df_pg1.boxplot(column=['AP'] ,figsize=(8,5))
```

```python
df_pg1.boxplot(column=['PE'] ,figsize=(8,5))
```
